```python
plt.figure(figsize=(8, 6))
plt.plot(lg_fpr, lg_tpr, label=f'Logistic Regression (AUC = {log_reggr_auc:.2f})')
plt.plot(svm_fpr, svm_tpr, label=f'SVM (AUC = {svm_auc:.2f})')
plt.plot(knn_fpr, knn_tpr, label=f'KNN (AUC = {knn_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')

plt.title('ROC Curve of all Machine Learning Models')
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.legend(loc='lower right')
plt.grid(alpha=0.4)
plt.show()
```

```python
#pre tuning and post tuning
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score

base_svm = Pipeline([
    ('standardscaler', StandardScaler()),
    ('Support Vector Machine', SVC(probability=True))
])


base_svm.fit(X_train, y_train)
base_svm_pred = base_svm.predict(X_test)
base_svm_prob = base_svm.predict_proba(X_test)[:, 1]

print("(Support Vector Machine (Pre-Tuning)")
print("Accuracy:", accuracy_score(y_test, base_svm_pred))
print("ROC AUC:", roc_auc_score(y_test, base_svm_prob))
print(classification_report(y_test, base_svm_pred))
```

```python
from sklearn.model_selection import GridSearchCV

svm_parameters = {
    'Support Vector Machine__C': [0.01, 0.1, 1, 10, 100, 1000, 10000],
    'Support Vector Machine__gamma': [0.001, 0.01, 0.1, 1.0, 10.0, 'scale', 'auto'],
    'Support Vector Machine__kernel': ['linear', 'rbf', 'poly','sigmoid']
}

svm_pipeline = Pipeline([
    ('standardscaler', StandardScaler()),
    ('Support Vector Machine', SVC(probability=True))
])

grid_svm = GridSearchCV(svm_pipeline, svm_parameters, cv=5, scoring='roc_auc', n_jobs=-1)
grid_svm.fit(X_train, y_train)


best_svm = grid_svm.best_estimator_
tuned_svm_pred = best_svm.predict(X_test)
tuned_svm_prob = best_svm.predict_proba(X_test)[:, 1]

print("(Support Vector Machine (Post-Tuning)")
print("Best Parameters:", grid_svm.best_params_)
print("Accuracy:", accuracy_score(y_test, tuned_svm_pred))
print("ROC AUC:", roc_auc_score(y_test, tuned_svm_prob))
print(classification_report(y_test, tuned_svm_pred))
```
